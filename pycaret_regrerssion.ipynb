```python
import pandas as pd
df = pd.read_csv('day.csv')
df.head()
```

```python
from pycaret.regression import setup, compare_models, evaluate_model, save_model, load_model, predict_model
```

```python
s = setup(data = df, target = 'cnt')
```

```python
best_model = compare_models()
```

```python
evaluate_model(best_model)
```

```python
test = pd.read_csv('day_test.csv')
predict_model(best_model, data=test)
```

```python
save_model(best_model, 'best_model')
```

```python
new_model = load_model('best_model')
```
